```python
def get_test_space(space):
    w = int(math.sqrt(len(space[0])))
    test_space = np.zeros(shape=(space.shape[0], w, w))
    for i in range(len(space)):
        temp = []
        for c in space[i]:
            temp.append(c)

        test_space[i] = np.asarray(temp).reshape(w, w)
    return test_space
```

```python
# -*- coding: utf-8 -*-
# @Time : 2020/8/7 下午2:44
# @Author : cmk
# @File : plot_ACCA.py

import numpy as np
from utils.utils import *
import matplotlib.pyplot as plt
from tqdm import tqdm


def get_space():
    datas = read_data("./temp.pkl")
    datas = datas[:100]
    datas = [datas[i] for i in range(0, len(datas), 2)]
    space = [[] for _ in range(len(datas))]

    colors_dict = {
        '00': 'white',
        '10': 'black',
        '01': 'lightgray',
        '02': 'lightgray',
        '11': 'darkgray',
        '12': 'darkgray'
    }

    for i in range(len(datas)):
        for data in datas[i]:
            cell = data['cell']
            space[i].append(colors_dict[cell['curr'] + str(cell['t'])])
    return np.asarray(space)


def get_space2():
    datas = read_data("./temp.pkl")
    # datas = datas[:100]
    datas = [datas[i] for i in range(0, len(datas), 2)]
    space = [[] for _ in range(len(datas))]

    colors_dict = {
        # '00': 255,
        # '10': 0,
        # '01': 150,
        # '02': 150,
        # '11': 50,
        # '12': 50
        '00': 0,
        '10': 1,
        '01': 0.3,
        '02': 0.3,
        '11': 0.7,
        '12': 0.7
    }

    for i in range(len(datas)):
        for data in datas[i]:
            cell = data['cell']
            space[i].append(colors_dict[cell['curr'] + str(cell['t'])])
    return np.asarray(space)



```
